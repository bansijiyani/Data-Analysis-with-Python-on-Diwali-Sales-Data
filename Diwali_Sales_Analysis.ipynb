# Exploratory Data Analysis

### Gender

```python
import matplotlib.pyplot as plt

sales_gen = df.groupby(['Gender'])['Amount'].sum().reset_index()

plt.bar(sales_gen['Gender'], sales_gen['Amount'])
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.title('Total Amount by Gender')
plt.show()
```

<h3>Conclusion </h3><h4>From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men</h4>

### Age

```python
grouped_data = df.groupby(['Age Group', 'Gender']).size().unstack()

# Plotting count of males and females by age group
grouped_data.plot(kind='bar',stacked=True)

# Set labels and title
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Count of Male and Female by Age Group')

plt.xticks(rotation=45)

# Display plot
plt.legend(title='Gender')
plt.show()
```

```python
# Grouping the data by 'Age Group' and calculating the total 'Amount' for each group
sales_age = df.groupby(['Age Group'])['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)

# Plotting the bar chart
plt.bar(sales_age['Age Group'], sales_age['Amount'])
plt.xlabel('Age Group')
plt.ylabel('Total Amount')
plt.title('Total Amount vs Age Group')
plt.show()
```

<h4>From above graphs we can see that most of the buyers are of age group between 26-35 yrs female</h4>

### State

```python
# Grouping the data by 'State' and calculating the total 'Orders' for each state
sales_state = df.groupby(['State'])['Orders'].sum().reset_index().sort_values(by='Orders', ascending=False).head(10)

# Plotting the line chart
plt.plot(sales_state['State'], sales_state['Orders'], marker='o')
plt.xlabel('State')
plt.ylabel('Total Orders')
plt.title('Total Orders from Top 10 States using Line Chart')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

<h4>From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively</h4>


### Marital Status

```python
# Counting the occurrences of each category in the 'Marital_Status' column
marital_status_counts = df['Marital_Status'].value_counts()

# Plotting the scatter plot
plt.scatter(marital_status_counts.index, marital_status_counts.values)

# Adding labels to the points
for i, (category, count) in enumerate(zip(marital_status_counts.index, marital_status_counts.values)):
    plt.text(category, count, str(count), ha='center', va='bottom')

# Setting labels and title
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.title('Count of Marital Status')

plt.show()
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

### Occupation

```python
import matplotlib.pyplot as plt

# Counting the occurrences of each category in the 'Occupation' column
occupation_counts = df['Occupation'].value_counts()

# Plotting the histogram (bar plot)
plt.bar(occupation_counts.index, occupation_counts.values,color = 'red')

# Adding labels to the bars
for i, count in enumerate(occupation_counts.values):
    plt.text(i, count, str(count), ha='center', va='bottom')

# Setting labels and title
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title('Count of Occupations')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

plt.show()

# print(occupation_counts.values) #it return count of occupation
```

```python
import matplotlib.pyplot as plt

# Grouping the data by 'Occupation' and calculating the total 'Amount' for each group
sales_state = df.groupby(['Occupation'])['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)

# Plotting the bar plot
plt.figure(figsize=(10, 5))
plt.bar(sales_state['Occupation'], sales_state['Amount'],color = 'cyan')

# Setting labels and title
plt.xlabel('Occupation')
plt.ylabel('Total Amount')
plt.title('Total Amount by Occupation')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

plt.show()
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Product Category

```python
# Counting the occurrences of each category in the 'Product_Category' column
product_category_counts = df['Product_Category'].value_counts()

# Plotting the line chart
plt.figure(figsize=(15, 5))
plt.plot(product_category_counts.index, product_category_counts.values, marker='o', linestyle='-')

# Adding labels to the points
for i, count in enumerate(product_category_counts.values):
    plt.text(i, count, str(count), ha='center', va='bottom')

# Setting labels and title
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.title('Count of Product Categories')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

plt.grid(True)

plt.show()
```

```python
# Grouping the data by 'Product_Category' and calculating the total 'Amount' for each group
sales_state = df.groupby(['Product_Category'])['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False).head(10)

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sales_state['Amount'], labels=sales_state['Product_Category'], autopct='%1.1f%%', startangle=140)

# Setting title
plt.title('Top 10 Product Categories by Total Amount')

# Adding a legend with labels
plt.legend(sales_state['Product_Category'], loc="lower right")

plt.show()
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
# Grouping the data by 'Product_ID' and calculating the total 'Orders' for each group
sales_state = df.groupby(['Product_ID'])['Orders'].sum().reset_index().sort_values(by='Orders', ascending=False).head(10)

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sales_state['Orders'], labels=sales_state['Product_ID'], autopct='%1.1f%%', startangle=140)

# Setting title
plt.title('Top 10 Product IDs by Total Orders')
plt.legend(sales_state['Product_ID'], loc="upper left")

plt.show()
```
